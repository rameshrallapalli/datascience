```python
train = pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
test = pd.read_csv('/kaggle/input/digit-recognizer/test.csv')
sample = pd.read_csv('/kaggle/input/digit-recognizer/sample_submission.csv')
```

```python
train.isnull().sum().sum(), test.isnull().sum().sum()
```

```python
ImageId = test.index + 1

y = train.iloc[:,0]
X = train.iloc[::,1:]
X_test = test
```

```python
#the MNIST dataset has images with pixel values in the range [0, 255]
X = X / 255
X_test = X_test / 255
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.1, random_state=1)
X_train.shape, X_val.shape, y_train.shape, y_val.shape, X_test.shape
```

```python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(hidden_layer_sizes=(50, ), max_iter=50, alpha=1e-4, solver='sgd', verbose=10, random_state=1, learning_rate_init=.1).fit(X_train, y_train)
print(model.score(X_train, y_train))
```

```python
y_pred = model.predict(X_val)
print(model.score(X_val, y_val))
```

```python
df_val=pd.DataFrame({'actual_value': y_val, 'predicted_value':y_pred})
df_val.reset_index(drop=True, inplace=True)
df_val
```

```python
prediction = model.predict(X_test)
prediction
```

```python
output = pd.DataFrame({'ImageId': ImageId, 'Label': prediction})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
